```python
import pandas as pd

# Assuming the dataset is stored in a variable named df

# Create batting sub-dataset
batting_df = df[['match_id', 'over', 'ball', 'runs', 'isBoundary', 'isWide', 'isNoball',
                 'batsman1_id', 'batsman1_name', 'batsman1_runs', 'batsman1_balls',
                 'batsman2_id', 'batsman2_name', 'batsman2_runs', 'batsman2_balls']].copy()

# Create bowling sub-dataset
bowling_df = df[['match_id', 'over', 'ball', 'runs', 'bowler1_id', 'bowler1_name', 
                 'bowler1_overs', 'bowler1_maidens', 'bowler1_runs', 'bowler1_wkts',
                 'bowler2_id', 'bowler2_name', 'bowler2_overs', 'bowler2_maidens',
                 'bowler2_runs', 'bowler2_wkts']].copy()

# Calculate batting average and strike rate for each batsman
batsman1_stats = batting_df.groupby('batsman1_name').agg(
    matches_played=('match_id', 'nunique'),
    runs_scored=('batsman1_runs', 'sum'),
    balls_faced=('batsman1_balls', 'sum'),
    batting_average=('batsman1_runs', 'mean'),
    strike_rate=('batsman1_runs', lambda x: (sum(x) / sum(batting_df['batsman1_balls'])) * 100)
).reset_index()

batsman2_stats = batting_df.groupby('batsman2_name').agg(
    matches_played=('match_id', 'nunique'),
    runs_scored=('batsman2_runs', 'sum'),
    balls_faced=('batsman2_balls', 'sum'),
    batting_average=('batsman2_runs', 'mean'),
    strike_rate=('batsman2_runs', lambda x: (sum(x) / sum(batting_df['batsman2_balls'])) * 100)
).reset_index()

# Calculate bowling average and economy rate for each bowler
bowler1_stats = bowling_df.groupby('bowler1_name').agg(
    matches_played=('match_id', 'nunique'),
    runs_conceded=('bowler1_runs', 'sum'),
    balls_bowled=('bowler1_overs', 'sum'),
    wickets_taken=('bowler1_wkts', 'sum'),
    bowling_average=('bowler1_runs', lambda x: x.sum() / x.count() if x.count() > 0 else 0),
    economy_rate=('bowler1_runs', lambda x: (x.sum() / (int(x.sum() / x.count()) * 6)) if x.count() > 0 else 0)
).reset_index()

bowler2_stats = bowling_df.groupby('bowler2_name').agg(
    matches_played=('match_id', 'nunique'),
    runs_conceded=('bowler2_runs', 'sum'),
    balls_bowled=('bowler2_overs', 'sum'),
    wickets_taken=('bowler2_wkts', 'sum'),
    bowling_average=('bowler2_runs', lambda x: x.sum() / x.count() if x.count() > 0 else 0),
    economy_rate=('bowler2_runs', lambda x: (x.sum() / (int(x.sum() / x.count()) * 6)) if x.count() > 0 else 0)
).reset_index()


```

```python
# Merge statistics for batsman1 and batsman2
batsmen_stats = pd.concat([batsman1_stats, batsman2_stats], ignore_index=True)

# Group by batsman name and aggregate statistics
batsmen_stats = batsmen_stats.groupby('batsman1_name').agg(
    matches_played=('matches_played', 'sum'),
    runs_scored=('runs_scored', 'sum'),
    balls_faced=('balls_faced', 'sum'),
    batting_average=('batting_average', 'mean'),
    strike_rate=('strike_rate', 'mean')
).reset_index()
```

```python
# Merge statistics for bowler1 and bowler2
bowlers_stats = pd.concat([bowler1_stats, bowler2_stats], ignore_index=True)

# Group by bowler name and aggregate statistics
bowlers_stats = bowlers_stats.groupby('bowler1_name').agg(
    matches_played=('matches_played', 'sum'),
    runs_conceded=('runs_conceded', 'sum'),
    balls_bowled=('balls_bowled', 'sum'),
    wickets_taken=('wickets_taken', 'sum'),
    bowling_average=('bowling_average', 'mean'),
    economy_rate=('economy_rate', 'mean')
).reset_index()
```

# Data Analysis: you can use different methods to solve the problem you proposed, and find out the best solution.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# Define features and target variable
features = ['matches_played', 'balls_faced', 'strike_rate']
target = 'batting_average'

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(batsmen_stats[features], batsmen_stats[target], test_size=0.2, random_state=42)

# Initialize and train Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict runs scored by batsmen on the testing set using Linear Regression
lr_y_pred = lr_model.predict(X_test)

# Calculate Mean Squared Error for Linear Regression
lr_mse = mean_squared_error(y_test, lr_y_pred)
print("Linear Regression Mean Squared Error:", lr_mse)

# Initialize and train Random Forest model
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)

# Predict runs scored by batsmen on the testing set using Random Forest
rf_y_pred = rf_model.predict(X_test)

# Calculate Mean Squared Error for Random Forest
rf_mse = mean_squared_error(y_test, rf_y_pred)
print("Random Forest Mean Squared Error:", rf_mse)

# Initialize and train Decision Tree model
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)

# Predict runs scored by batsmen on the testing set using Decision Tree
dt_y_pred = dt_model.predict(X_test)

# Calculate Mean Squared Error for Decision Tree
dt_mse = mean_squared_error(y_test, dt_y_pred)
print("Decision Tree Mean Squared Error:", dt_mse)
```
